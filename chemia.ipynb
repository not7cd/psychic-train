```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (8,8)
```

```python
con = [0.07, 0.13, 0.2, 0.27, 0.33]
log_con = np.log(con)
time = [85., 43., 29., 23., 17.]


k, b = np.polyfit(time, log_con, 1)
print(k, b, log_con)
```

```python
fig = plt.figure()
ax = fig.gca()
# ax.set_yscale('log')

# plt.xlabel('odrotność temperatury [1/K]')
# plt.ylabel('log ciśnienie [ln(hPa)]')
plt.errorbar(y=log_con, x=time, yerr=0., xerr=1, fmt='.')
plt.plot(time, np.add(np.multiply(k, time), b), '--')


plt.grid()
plt.show()
```

```python
sm = [10.47, 184.6, 385, 528.8, 633.5, 708.8, 764.0, 803.3, 831.9, 852.0, 868.0]
sm_t = list(range(0,21,2))
sm_diff = np.subtract(910.7, sm)
# print(sm)
# print(sm_diff)
# print(sm_t)
for t, d in zip(sm_t, np.log(sm_diff)):
    print('t:{}  s:{:.3f}'.format(t, d))
```

```python
fig = plt.figure()
ax = fig.gca()
ax.set_yscale('log')

# plt.xlabel('odrotność temperatury [1/K]')
# plt.ylabel('log ciśnienie [ln(hPa)]')
plt.errorbar(y=sm_diff, x=sm_t, yerr=0., xerr=0.04, fmt='.')

plt.grid()
plt.show()
```
